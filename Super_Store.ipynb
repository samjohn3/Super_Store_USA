Analyze the profit by Region

```python
profit_region = df.groupby("Region")["Profit"].sum().sort_values()
```

```python
profit_region
```

Visualization

```python
sns.barplot(
    x=profit_region.index,
    y=profit_region.values,
    hue=profit_region.index,   # Assign x to hue
    palette="viridis",
    legend=False               # Hide legend
)
plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```

Analyze Monthly Sales Trend

```python
df["Month"] = df["Order Date"].dt.to_period("M")
sales_monthly = df.groupby("Month")["Sales"].sum()
sales_monthly.plot(kind="line")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True) #grid for better readability
plt.show()
```
